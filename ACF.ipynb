```python
import pandas as pd

def calculate_cash_flow_a(df):
    # Ensure all necessary columns are present
    required_columns = [
        'Profit/(Loss) - Starting Line - Cash Flow',
        'Non-cash Items & Reconciliation Adjustments - Cash Flow',
        'Other Non-Cash Items & Reconciliation Adjustments - Cash Flow - to Reconcile',
        'Depreciation, Depletion & Amortization including Impairment - Cash Flow - to Reconcile',
        'Assets Sale - Gain/(Loss) - Cash Flow - to Reconcile',
        'Working Capital - Increase/(Decrease) - Cash Flow'
    ]
    
    # Check if all required columns are present in the DataFrame
    if not set(required_columns).issubset(df.columns):
        raise ValueError("Missing required columns in the DataFrame")

    # Calculate Cash Flow A
    df['Cash Flow A'] = (
        df['Profit/(Loss) - Starting Line - Cash Flow'] +
        df['Non-cash Items & Reconciliation Adjustments - Cash Flow'] +
        df['Other Non-Cash Items & Reconciliation Adjustments - Cash Flow - to Reconcile'] +
        df['Depreciation, Depletion & Amortization including Impairment - Cash Flow - to Reconcile'] +
        df['Assets Sale - Gain/(Loss) - Cash Flow - to Reconcile'] +
        df['Working Capital - Increase/(Decrease) - Cash Flow']
    )
    
    return df
```

```python
import pandas as pd

def estimate_discount_rate(df):
    # Extract necessary variables
    interest_paid = df['Interest Paid - Cash Flow - Supplemental'].iloc[0]
    debt_issued = df['Debt - Long-Term & Short-Term - Issuance/(Retirement) - Total - Cash Flow'].iloc[0]
    
    # Check for NaN values or zero debt issued
    if pd.isna(interest_paid) or pd.isna(debt_issued):
        raise ValueError("Interest Paid or Debt Issued contains NaN values.")
    
    if debt_issued == 0:
        raise ValueError("Debt Issued is zero, cannot calculate cost of debt.")
    
    # Calculate cost of debt (discount rate)
    cost_of_debt = interest_paid / debt_issued
    
    return cost_of_debt

# Call the function with your actual df
try:
    discount_rate = estimate_discount_rate(df)
    print(f"Estimated Discount Rate: {discount_rate:.4f}")
except ValueError as e:
    print(e)
```

```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import seaborn as sns

# Download data for Merck KGaA and DAX 30 Index (last 8 years)
start_date = '2014-07-01'
end_date = '2024-07-01'

# Download monthly adjusted closing prices
merck = yf.download('MRK.DE', start=start_date, end=end_date, interval='1mo')
dax = yf.download('^GDAXI', start=start_date, end=end_date, interval='1mo')

# Keep only the Adjusted Closing Prices
merck_close = merck['Adj Close']
dax_close = dax['Adj Close']

# Print to check the indices of both Series
print("Merck Indices:")
print(merck_close.index)
print("\nDAX Indices:")
print(dax_close.index)

# Align the indices using an inner join (keep only dates common to both Series)
df = pd.concat([merck_close, dax_close], axis=1, join='inner')
df.columns = ['Merck', 'DAX']

# Drop any rows with missing values
df.dropna(inplace=True)

print("\nCombined Data:")
print(df)

# Calculate monthly returns
df['Merck_Return'] = df['Merck'].pct_change()
df['DAX_Return'] = df['DAX'].pct_change()
df.dropna(inplace=True)

# Calculate Beta
cov_matrix = np.cov(df['Merck_Return'], df['DAX_Return'])
beta = cov_matrix[0, 1] / cov_matrix[1, 1]
print(f"\nBeta of Merck KGaA relative to the DAX 30 Index (last 10 years): {beta:.4f}")

# Plot the scatter plot of returns with a regression line
sns.regplot(x='DAX_Return', y='Merck_Return', data=df, ci=None, line_kws={"color": "red"})
plt.title('Merck KGaA vs. DAX 30 Index Monthly Returns (Last 10 Years)')
plt.xlabel('DAX 30 Index Monthly Returns')
plt.ylabel('Merck KGaA Monthly Returns')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import yfinance as yf

# Download historical market data (DAX 30 Index as an example)
ticker = "^GDAXI"  # DAX 30 Index ticker
start_date = "2014-07-01"
end_date = "2024-07-01"

# Fetch monthly data for the last 10 years
market_data = yf.download(ticker, start=start_date, end=end_date, interval='1mo')

# Calculate monthly returns
market_data['Monthly_Return'] = market_data['Adj Close'].pct_change()

# Drop NaN values
market_data.dropna(inplace=True)

# Calculate the average monthly return and annualize it
average_monthly_return = market_data['Monthly_Return'].mean()
annualized_return = (1 + average_monthly_return) ** 12 - 1

print(f"Average Annual Market Return: {annualized_return:.2%}")
```

```python
# Given inputs
risk_free_rate = 2.14 / 100     # Convert 2.14% to decimal
beta = 0.7398                   # Beta of the stock
market_return = 8.54 / 100      # Convert 8.54% to decimal

# Calculate market risk premium
market_risk_premium = market_return - risk_free_rate

# Calculate cost of equity using CAPM
cost_of_equity = risk_free_rate + beta * market_risk_premium

# Display the result
print(f"Cost of Equity (k_S): {cost_of_equity:.2%}")
```

## 📊 Interpretation of the Cost of Equity

### Key Results:
- **Cost of Equity (\(k_S\))**: **6.91%**

### What Does This Mean?

1. **Definition**:
   - The **Cost of Equity** is the **required rate of return** that investors expect for holding the company's stock.
   - It reflects the compensation investors need for taking on the risk associated with the stock.

2. **Components**:
   - **Risk-Free Rate (\(R_f\))**: Represents a **risk-free investment** (e.g., a 10-year government bond).
   - **Beta (\(\beta\))**: Measures the **stock's volatility** relative to the market.
   - **Market Risk Premium (\(R_m - R_f\))**: The additional return investors expect for investing in the market over a risk-free asset.

3. **Interpretation of 6.91%**:
   - Investors require a **6.91% annual return** for investing in this stock.
   - This return compensates them for the **time value of money** and the **risk** of the investment.
   - The cost of equity is used as the **discount rate** for evaluating future cash flows in valuation models like the **Discounted Cash Flow (DCF)**.

4. **Implications for Valuation**:
   - If the company's expected returns are **higher** than 6.91%, the stock may be considered **undervalued**.
   - If the expected returns are **lower**, the stock may be considered **overvalued**.

5. **Risk Perspective**:
   - A **beta of 0.7398** indicates that the stock is **less volatile** than the overall market.
   - The stock is likely to experience **smaller fluctuations** compared to the market index.

```python
# Define the given inputs again for clarity and completeness
import numpy as np

# Inputs
Re = 6.87 / 100  # Cost of equity in decimal
D = 9864  # Debt value (€)
E = 62651  # Equity value (€)
V = D + E  # Total value (€)
t = 29.9 / 100  # Tax rate in decimal
Dr = 2.7 / 100  # Updated cost of debt in decimal
terminal_growth_rate = 8 / 100  # Terminal growth rate in decimal
projection_period = 5  # Number of years for projection

# Free cash flows for each year (2024 to 2028)
free_cash_flows = [2882.175, 2820.206, 2753.506, 2681.715, 4489.806]  # (€)
terminal_value = 74830.11  # Terminal value (€)

# WACC formula
WACC = (E / V) * Re + (D / V) * Dr * (1 - t)

# Discount the free cash flows
discounted_fcffs = [
    fcff / ((1 + WACC) ** (i + 1)) for i, fcff in enumerate(free_cash_flows)
]

# Discount the terminal value
discounted_terminal_value = terminal_value / ((1 + WACC) ** projection_period)

# Calculate the total firm value
firm_value = sum(discounted_fcffs) + discounted_terminal_value

# Convert WACC to percentage for the final output
WACC_percentage = WACC * 100

# Print the results
{
    "WACC (%)": WACC_percentage,
    "Discounted Free Cash Flows (€)": discounted_fcffs,
    "Discounted Terminal Value (€)": discounted_terminal_value,
    "Firm Value (€)": firm_value,
}
```
